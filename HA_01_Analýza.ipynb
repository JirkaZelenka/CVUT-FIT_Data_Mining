## 2_Pocet_stran_a_kandidatu_2002-2018

```python
uniq_strany = vysledky.Strana.unique()  
uniq_kandidati = kandidati.Jméno.unique()

display(len(uniq_strany))
display(uniq_strany)
display(len(uniq_kandidati))

########################################################
# 41 stran, 1735 kandidátů, unikátních
```

```python
# Unikátní strany a kandidáti v jednotlivých letech

roky = [2002, 2006, 2010, 2014, 2018]
počet_stran_rok=[]
počet_kandidátů_rok = []

for i in range(5):
    počet_stran_rok.append(len(vysledky[vysledky["Rok"]== roky[i]].Strana.unique()))
    počet_kandidátů_rok.append(len(kandidati[kandidati["Rok_kandidatury"]== roky[i]].Jméno.unique()))
        
display(počet_stran_rok)
počet_kandidátů_rok
```

#### Zajimavost-korelace_0.82

```python
# Výpočet korelace

data = pd.DataFrame(list(zip(počet_stran_rok, počet_kandidátů_rok)))
data

corr = data.corr()
display(corr)

# Korelace 0.82 !!


ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

# Korelace počtu strana počtu kandidátů je nepřekvapivě vysoká, 0.82
```

### 2.2_Pocet_kandidatu_podle_stran

```python
########################################################
# Trochu neobratný kod, nezbyl čas na předělání
########################################################

vývoj_kandidátů_stran = DataFrame(uniq_strany)

dva =[]
šest = []
deset = []
čtrnáct = []
osmnáct = []

for i in range(len(vývoj_kandidátů_stran[0])):
    dva.append(kandidati[(kandidati['Rok_kandidatury'] == 2002) & (kandidati["Strana"] == vývoj_kandidátů_stran[0][i])].count()[0])
    šest.append(kandidati[(kandidati['Rok_kandidatury'] == 2006) & (kandidati["Strana"] == vývoj_kandidátů_stran[0][i])].count()[0])
    deset.append(kandidati[(kandidati['Rok_kandidatury'] == 2010) & (kandidati["Strana"] == vývoj_kandidátů_stran[0][i])].count()[0])
    čtrnáct.append(kandidati[(kandidati['Rok_kandidatury'] == 2014) & (kandidati["Strana"] == vývoj_kandidátů_stran[0][i])].count()[0])
    osmnáct.append(kandidati[(kandidati['Rok_kandidatury'] == 2018) & (kandidati["Strana"] == vývoj_kandidátů_stran[0][i])].count()[0])

    #Pro daný rok a danou stranu ze seznamu unikátních stran spočítej počet prvků, resp jmen ([0]) a prřidej je do sloupce pro daný rok
    
vývoj_kandidátů_stran["2002"] = dva
vývoj_kandidátů_stran["2006"] = šest
vývoj_kandidátů_stran["2010"] = deset
vývoj_kandidátů_stran["2014"] = čtrnáct
vývoj_kandidátů_stran["2018"] = osmnáct

vývoj_kandidátů_stran.head()
```

```python
#Vzhledem k velmému počtu stran, které se zúčastnily jediných voleb (36 ze 41 stran) se zde budu věnovat více pouze pěti stranám s více než jednou účastí:
#---------------------
# 5x: KSČM, ODS, ČSSD
# 4x: SMM
# 2x: ANO


#-------------------------------------------------------------
ČSSD_kandidáti_počet = vývoj_kandidátů_stran.iloc[9,1:]
ODS_kandidáti_počet = vývoj_kandidátů_stran.iloc[3,1:]
KSČM_kandidáti_počet = vývoj_kandidátů_stran.iloc[2,1:]
SMM_kandidáti_počet = vývoj_kandidátů_stran.iloc[15,1:]
ANO_kandidáti_počet = vývoj_kandidátů_stran.iloc[33,1:]
```

#### 2.2.1_Pocet_kandidatu_v_peti_nejcastejsich_stranach

```python

df = pd.DataFrame({'x': (2002,2006,2010,2014,2018), 'ČSSD': ČSSD_kandidáti_počet, "ODS": ODS_kandidáti_počet, "KSČM": KSČM_kandidáti_počet, "SMM": SMM_kandidáti_počet, "ANO": ANO_kandidáti_počet })
 
plt.style.use('fivethirtyeight')

plt.plot(df['x'], df['ČSSD'], color=oranžová)
plt.plot(df['x'], df['ODS'],  color=modrá) 
plt.plot(df['x'], df['KSČM'], color=červená) 
plt.plot(df['x'], df['SMM'],  color=zelená) 
plt.plot(df['x'], df['ANO'], color=tyrkysová) 
    
plt.ylim([42,46])

plt.xticks(np.arange(2002, 2026, 4.0),rotation = 45)
plt.yticks(np.arange(42, 47, 1.0))

plt.title("Počet kandidátů v pěti nejčastějších stranách", loc='right', fontsize=18, fontweight=0, color=černá)
plt.xlabel("Rok")
plt.ylabel("Kandidáti")
plt.legend(loc=[0,0], bbox_to_anchor=(0.9,0.3))
```
