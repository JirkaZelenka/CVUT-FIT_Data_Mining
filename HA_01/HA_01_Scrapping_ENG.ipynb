```python
years = [2002, 2006, 2010, 2014, 2018]

df_people = []

for i in years:
    #Looping is done through years, not index + indexed years
    
    url = r"https://www.volby.cz/pls/kv" + str(i) + "/kv21111?xjazyk=CZ&xid=0&xv=11&xdz=3&xnumnuts=4205&xobec=567027&xstrana=0"
    table = pd.read_html(url)
    table[0]= table[0].drop(table[0].index[0:2])      # Two header rows deleted  
    table[0][12] = i                                  # Adding a year of candidature
    
#print(table[0])
    df_people.append(table[0])

#########################################
# Total Candidates:
# 365 + 415 + 394 + 547 + 514 = 2235
```
